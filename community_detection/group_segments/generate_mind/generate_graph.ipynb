# Hierarchical Community.

```python
import networkx as nx
G_pruned = nx.read_gpickle('minds/se/se_ent_graph_wcosine_pruned_v3.gpkl')
```

```python
from community import best_partition

comm = best_partition(G_pruned, resolution=0.9)
```

```python
community_set_sorted = sorted(comm.items(), key=lambda kv:kv[1], reverse=False)
```

```python
clusters = []
temp = []
prev_com = 0
for index, (word, cluster) in enumerate(community_set_sorted):
    if prev_com == cluster:
        temp.append(word)
        if index == len(community_set_sorted) - 1:
            clusters.append(temp)
    else:
        clusters.append(temp)
        temp = []
        prev_com = cluster
        temp.append(word)
```

```python

community_set_collection = []
old_cluster = []
for cluster in clusters:
    if len(cluster) >= 2:
        subgraph_ent = G_pruned.subgraph(cluster)
        subgraph_com = best_partition(subgraph_ent, resolution=0.9)
        subgraph_com_sorted = sorted(subgraph_com.items(), key=lambda kv:kv[1], reverse=False)
        
        subgraph_cluster = []
        temp = []
        prev_com = 0
        for index, (word, cls) in enumerate(subgraph_com_sorted):
            if prev_com == cls:
                temp.append(word)
                if index == len(subgraph_com_sorted) - 1:
                    subgraph_cluster.append(temp)
            else:
                subgraph_cluster.append(temp)
                temp = []
                prev_com = cls
                temp.append(word)
        
        community_set_collection.extend(subgraph_cluster)
```

```python
com_map = {}
for cls, community in enumerate(community_set_collection):
    for ent in community:
        com_map[ent] = cls
```
